# 6. Outliers and Replacing them

```python
outliers=ds.quantile(q=(0.25,0.75))
outliers
```

```python
Q1 = ds.Age.quantile(0.25)
Q3 = ds.Age.quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
ds.median(numeric_only=True)
```

```python
ds['Age'] = np.where(ds['Age'] < lower_limit, 7, ds['Age'])
```

```python
sns.boxplot(x=ds.Age,showfliers = False)
```

# 7.Checking for Categorical columns and perform Encoding

```python
ds.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
ds.Sex = le.fit_transform(ds.Sex)

ds.head()
```

# 8. Split the data into dependent and independent variables

```python
y = ds["Sex"]
y.head()
```

```python
x=ds.drop(columns=["Sex"],axis=1)
x.head()
```

# 9.Scaling the values

```python
from sklearn.preprocessing import scale
X_Scaled = pd.DataFrame(scale(x), columns=x.columns)
X_Scaled.head()
```

# 10. Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X_Scaled, y, test_size=0.2, random_state=0)
```

# 11. Build the Model

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10,criterion='entropy')
```

```python
model.fit(X_Train,Y_Train)
```

```python
y_predict = model.predict(X_Test)
```

```python
y_predict_train = model.predict(X_Train)
```

# 12. Train the Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('Training accuracy: ',accuracy_score(Y_Train,y_predict_train))
```

# 13.Test the Model

```python
print('Testing accuracy: ',accuracy_score(Y_Test,y_predict))
```

# 14. Measure the performance using Metrics

```python
pd.crosstab(Y_Test,y_predict)
```

```python
print(classification_report(Y_Test,y_predict))
```
